```python
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
import random
import numpy as np
```

#### 样本生成

```python
# 设置每个类别的总数和准确率

class_num = 10
num_samples = [100, 100, 100, 100, 100, 100, 100, 100, 100, 100]
sam_np = np.array(num_samples)
# total_acc = 0.917 
# 定义每个类别的准确率
accuracy = [0.89, 0.92, 0.88, 0.93, 0.93, 0.90, 0.87, 0.79,  0.90, 0.91]
acc_np = np.array(accuracy)
total_acc = (sum(acc_np*sam_np))/(sum(sam_np))
print(total_acc)
# num_samples = [100, 100, 100, 100, 100, 100, 100, 100, 100, 100, 100, 100]
# accuracy = [0.9, 0.4, 0.6, 0.5, 0.7, 0.2, 0.8, 0.9, 0.65, 0.87, 0.69, 0.92]

# 生成样本x和预测结果y
x = []
y = []
for i in range(class_num):
    for j in range(num_samples[i]):
        x.append(i)
        if random.random() < accuracy[i]:
            y.append(i)
        else:
            y.append(random.choice([x for x in range(class_num) if x != i]))
# 输出样本和预测结果的长度，用于检查是否生成正确
print(len(x))
print(len(y))
```
